```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_path = 'A_to_G_S18.xlsx'  # Update with your file path
data = pd.ExcelFile(file_path)

# Parse the relevant sheet
sheet_data = data.parse('S18_A_to_G')

# Extract the 'a-to-g(%)' column
a_to_g_data = sheet_data['a-to-g(%)']

# Define bins for the range [0.0-0.1], [0.1-0.2], ..., [0.9-1.0]
bins = [i*10 for i in range(12)]

# Categorize data into bins and count occurrences in each bin
categorized_data = pd.cut(a_to_g_data, bins=bins, right=False, include_lowest=True)
counts = categorized_data.value_counts(sort=False)
print(counts)
print(sum(counts))

# Plot the bar graph
plt.figure(figsize=(10, 6))
plt.bar(counts.index.astype(str), counts.values, width=1, align='center', edgecolor='black')
plt.xlabel('Range of a-to-g(%)')
plt.ylabel('Count')
plt.title('Distribution of a-to-g(%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_path = './A_to_G_del_0_0.1_93.xlsx'
data = pd.ExcelFile(file_path)

# Load the first sheet
df = data.parse('Sheet1')

# Assuming the range of interest is the column 'a-to-g(%)'
column_of_interest = 'a-to-g(%)'

# Define the new range intervals for [0,1), [1,2), ..., [n, n+1)
bins_new = range(0, int(df[column_of_interest].max()) + 2, 1)  # Dynamic upper limit based on data
labels_new = [f"[{i},{i+1})" for i in bins_new[:-1]]

# Create a histogram by binning the data
df['Range_New'] = pd.cut(df[column_of_interest], bins=bins_new, labels=labels_new, right=False)
range_counts_new = df['Range_New'].value_counts().sort_index()

# Print the counts for each range
counts_df = range_counts_new.reset_index(name='Count').rename(columns={'index': 'Range'})
print(counts_df)

# Plot the bar graph
plt.figure(figsize=(10, 6))
range_counts_new.plot(kind='bar')
plt.xlabel('Range of a-to-g(%)')
plt.ylabel('Count')
plt.title('Distribution of a-to-g(%) in Ranges of 1%')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_path = './A_to_G_S18.xlsx'
data = pd.ExcelFile(file_path)

# Load the sheet into a DataFrame
df = data.parse('S18_A_to_G')

# Extract the column of interest for range counting
column_name = 'a-to-g(%)'
values = df[column_name]

# Create bins and count values in each range
bins = range(0, 102, 1)  # Create bins from 0 to 100 with step 1
hist, bin_edges = pd.cut(values, bins=bins, right=False, retbins=True, labels=False)

# Count the occurrences of each bin
bin_counts = pd.value_counts(hist, sort=True)
print(bin_counts)
print(sum(bin_counts))
# Plot the bar graph
plt.figure(figsize=(12, 6))
plt.bar(bin_counts.index, bin_counts.values, width=1, edgecolor="black")
plt.title("Distribution of a-to-g(%) Values")
plt.xlabel("Ranges (e.g., [0,1), [1,2), etc.)")
plt.ylabel("Count")
plt.xticks(ticks=range(0, len(bins)-1, 10), 
           labels=[f"[{bins[i]},{bins[i+1]})" for i in range(0, len(bins)-1, 10)], 
           rotation=90)
plt.tight_layout()
plt.show()
```
